```python
import sys
sys.path.append("/home/nikolay/workspace/1.current/merlin_service_2/app")

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
from glob import glob
from tqdm.auto import tqdm

from merlin.face_detection import FaceToolbox
from merlin.glazomer.anfas_glazomer import calculate_features_anfas, Anfas_BrowForm_model

faceAnalizer = FaceToolbox(path='/home/nikolay/workspace/1.current/merlin_service_2/app')
browFrom_model = Anfas_BrowForm_model()
```

```python

path = '/home/nikolay/workspace/mfo/data/total'
folder = '04620-31001_5720201e-b163-11e2-b0f0-d4ae52c6cb41/'
def process_folder(folder, exceptions_count):
    dff = pd.DataFrame()
    dff['left_eye'] = None
    dff['right_eye'] = None
    dff['XFrontal'] = None
    dff['YFrontal'] = None
    anfas_glazomer = {}
    
    for i, file in enumerate(os.listdir(os.path.join(path, folder))):
        if not file.endswith('txt'):
            try:
                img_path = os.path.join(path, folder, file)
                img = cv2.imread(img_path)
                img.shape

                faces = faceAnalizer.get_faces(img)
                if len(faces) == 0:
                    continue

                face = faces[0]
                points = np.append(face['kps'][:,0], face['kps'][:,1])
                XFrontal, YFrontal = faceAnalizer.find_pose(points)
                landmarks68 = faceAnalizer.get_68_points(img, face)
                landmarks29 = faceAnalizer.get_29_anfas_points(img, face)
                anfas_glazomer = calculate_features_anfas(landmarks29, landmarks68, img, printout=False,
                                                          Anfas_BrowForm_model=browFrom_model)

                dff.loc[i, ['XFrontal', 'YFrontal', 'folder_name', 'left_eye', 'right_eye']] = XFrontal, YFrontal, folder, faces[0]['kps'][0].tolist(), faces[0]['kps'][1].tolist()

                pts_path_68 = '/'.join(img_path.split('/')[:-1]) + '/' + img_path.split('/')[-1][:-3] + '68pts'
                pts_path_29 = '/'.join(img_path.split('/')[:-1]) + '/' + img_path.split('/')[-1][:-3] + '29pts'
                np.savetxt(pts_path_68, landmarks68.astype('int'), fmt="%d")
                np.savetxt(pts_path_29, landmarks29.astype('int'), fmt="%d")
                
#                 return dff, anfas_glazomer

            except Exception as e:
                exceptions_count += 1
                pass
            
    return dff, anfas_glazomer

df = pd.DataFrame()

exceptions_count = 0 
for folder_num, folder in enumerate(tqdm(os.listdir(path)[:])):
    dff, anfas_glazomer = process_folder(folder, exceptions_count)
    if dff.shape[0] == 0:
        continue
    try:
        #remove all photos except the one with min Y angle
        good_rakurs_ind = dff[['XFrontal', 'YFrontal']].sum(axis=1).abs().idxmin()
        df = df.append(dff.loc[good_rakurs_ind], ignore_index=True)
        
        anfas_glazomer['feature_values'].pop('Форма бровей')        
        df.loc[df.index[-1], anfas_glazomer['feature_values'].keys()] = anfas_glazomer['feature_values'].values()
        
    
#         bad_photos = [dff.at[i, 'img_path'] for i in dff.index if i != good_rakurs.index[0]]
#         [os.remove(i) for i in bad_photos]


#         new_img_path = '.'.join(good_rakurs.at[good_index, 'img_path'].split('.')[:-1])+'.jpg'
#         if img_path != new_img_path:
#             os.remove(img_path)
#         cv2.imwrite(new_img_path, rotated)

    except Exception as exp:
        print(exp)
```
